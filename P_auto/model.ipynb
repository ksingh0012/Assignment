```python
import pandas as pd
```

```python
# Load dataset
data = pd.read_csv("data.csv")
data.head()
```

```python
data.shape
```

```python
# duplicate rows
num_dup = data.duplicated().sum()
print(f"number of duplicate rows: {num_dup}")
```

```python
# remove duplicate rows
data = data.drop_duplicates()
data.head(10)
```

```python
# after removing duplicate rows
data.shape
```

```python
# normalizing values
data['has_joined_event'] = data['has_joined_event'].str.lower().map({
    "yes" : True,
    "no" : False
})
print(data["has_joined_event"].unique())
```

```python
data.head()
```

```python
# flag missing and incomplete LinkedIn profiles
def is_incomplete_linkedin(profile):
    if pd.isna(profile):
        return True
    profile = profile.strip().lower()
    return not profile.startswith("http") or "linkedin.com" not in profile

data['linkedin_flag'] = data['What is your LinkedIn profile?'].apply(is_incomplete_linkedin)

data['job_title_missing'] = data['Job Title'].isna() | (data['Job Title'].str.strip() == '')

data.to_csv("cleaned_output.csv", index=False)
print("Done")
```

```python
import os
# Load the cleaned data
df = pd.read_csv('cleaned_output.csv')

# Extract necessary columns
df = df[['email', 'first_name', 'has_joined_event', 'Job Title', 'What is your LinkedIn profile?']].copy()
```

```python
# Message generation function
def generate_message(row):
    name = row['first_name'].strip() if pd.notna(row['first_name']) else "there"
    job = row['Job Title'].strip() if pd.notna(row['Job Title']) and row['Job Title'].strip() else "professional"
    joined = row['has_joined_event']

    if joined:
        return f"Hey {name}, thanks for joining our session! As a {job.lower()}, we think you’ll love our upcoming AI workflow tools. Want early access?"
    else:
        return f"Hi {name}, sorry we missed you at the last event! We’re preparing another session that might better suit your interests as a {job.lower()}."

# Apply the message logic
df['message'] = df.apply(generate_message, axis=1)
```

```python
# Save to CSV
df[['email', 'message']].to_csv('personalized_messages.csv', index=False)
```
